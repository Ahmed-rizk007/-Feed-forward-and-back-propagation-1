```python
import numpy as np
```

```python
x = np.array([0.05, 0.10])  
```

```python
# input -> hidden (2x2)
w1 = np.random.uniform(-0.5, 0.5, (2, 2))
```

```python
# hidden -> output (2x2)
w2 = np.random.uniform(-0.5, 0.5, (2, 2))
```

```python
# ===== Biases
b1 = 0.5
b2 = 0.7
```

```python
# ===== Activation Function (tanh)
def tanh(x):
    return np.tanh(x)
```

```python
# Hidden Layer
net_h = np.dot(x, w1) + b1
out_h = tanh(net_h)
```

```python
# Output Layer
net_o = np.dot(out_h, w2) + b2
out_o = tanh(net_o)
```

```python
print("Hidden Layer Output:")
print(out_h)

print("\nFinal Network Output:")
print(out_o)
```
